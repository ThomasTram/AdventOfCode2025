```python
# Day 1
# Part 1
with open('day1.txt') as fid:
    strings_data = fid.read().strip().split('\n')



numbers = [int(d.replace('L','-').replace('R','+')) for d in strings_data]
value = 50
count = 0
for num in numbers:
    value = (value + num) % 100
    if value == 0:
        count += 1
print(count)

# Part 2


value = 50
count = 0
last_negative = False
for num in numbers:
    if value == 0 and num >= 0 and last_negative:
        add_extra = 1
    elif value == 0 and num <= 0 and not last_negative:
        add_extra = -1
    else:
        add_extra = 0
    value += num
    crossings = value // 100
    value = value % 100
    count += abs(crossings) + add_extra
    last_negative = num <= 0

print(count)
```

```python
# Day 2
# Part 1

data='11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528,446443-446449,38593856-38593862,565653-565659,824824821-824824827,2121212118-2121212124'
all_ranges = data.split(',')

with open('day2.txt') as fid:
    all_ranges = fid.read().strip().split(',')
id_sum = 0
for r in all_ranges:
    beg, end = map(int, r.split('-'))
    for id in range(beg, end + 1):
        idstr = str(id)
        N = len(idstr)
        if N % 2:
            continue
        idstr1, idstr2 = idstr[:N//2], idstr[N//2:]
        if idstr1 == idstr2:
            id_sum += id
        
print(id_sum)

# Part 2
all_primes = (2, 3, 5, 7)
def is_invalid(idstr):
    N = len(idstr)
    for p in all_primes:
        if N % p:
            continue
        delta = N // p
        s1 = idstr[:delta]
        all_equal = all([s1 == idstr[delta*j:delta*(j+1)] for j in range(1, p)])
        if all_equal:
            return True
    return False
    
id_sum = 0
for r in all_ranges:
    beg, end = map(int, r.split('-'))
    for id in range(beg, end + 1):
        idstr = str(id)        
        if is_invalid(idstr):
            id_sum += id
print(id_sum)
```
