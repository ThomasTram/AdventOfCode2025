```python
# Day 1
# Part 1
with open('day1.txt') as fid:
    strings_data = fid.read().strip().split('\n')



numbers = [int(d.replace('L','-').replace('R','+')) for d in strings_data]
value = 50
count = 0
for num in numbers:
    value = (value + num) % 100
    if value == 0:
        count += 1
print(count)

# Part 2

value = 50
count = 0
last_negative = False
for num in numbers:
    if value == 0 and num >= 0 and last_negative:
        add_extra = 1
    elif value == 0 and num <= 0 and not last_negative:
        add_extra = -1
    else:
        add_extra = 0
    value += num
    crossings = value // 100
    value = value % 100
    count += abs(crossings) + add_extra
    last_negative = num <= 0

print(count)
```
